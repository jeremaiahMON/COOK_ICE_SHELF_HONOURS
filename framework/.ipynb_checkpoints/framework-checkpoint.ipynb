```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.features import rasterize
import geopandas as gpd
from rasterio.warp import reproject, Resampling
from pyproj import Transformer
```

## melt factor

### high BMR equals 1, low BMR equals 0

```python
infile = "CIS_BMR_mask_clipped_smooth.tif"
outfile = "bmr_factor_WAOM_high.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    bmr = src.read(1).astype(float)

    mask = np.isnan(bmr)

    valid = bmr[~mask]
    min_val, max_val = np.min(valid), np.max(valid)

    if min_val == max_val:
        norm = np.zeros_like(bmr, dtype=float)
    else:
        norm = (bmr - min_val) / (max_val - min_val)

    norm[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(norm.astype("float32"), 1)
```

```python
outfile = "bmr_factor_WAOM_high.tif"
with rasterio.open(outfile) as src:
    melt = src.read(1)  
    extent = (src.bounds.left,src.bounds.right,src.bounds.bottom,src.bounds.top,)  

plt.figure(figsize=(6, 4))
plt.imshow(melt, cmap="Reds", extent=extent, origin="upper", vmin = 0, vmax = 1)
plt.colorbar(label="BMR (0 = lowest BMRM 1 = highest BMR")
plt.title(" bmr_factor_WAOM_high ")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### inverted BMR scale

```python
infile = "bmr_factor_WAOM_high.tif"       
outfile = "bmr_factor_WAOM_low.tif"     

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    norm = src.read(1).astype(float)

    mask = np.isnan(norm)

    inv = 1 - norm

    inv[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(inv.astype("float32"), 1)
```

## crevassing factor

### inverting probability, NOT accounting for regions which are already crevassed

```python
infile = "MEDIAN_probability.tif"
outfile = "crevasse_factor_excluding_current_crevasses.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()  
    data = src.read(1)  

    # Invert probabilities: stability = 1 - probability
    stability = 1.0 - data

    if src.nodata is not None:
        stability[data == src.nodata] = src.nodata

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(stability, 1)

# This geotiff can not be used for the framework YET, because it does not account for regions which are ALREADY crevassed.
```

### accounting for regions which are already crevassed: CREVASSES = 0

```python
infile_raster = "crevasse_factor_excluding_current_crevasses.tif"
infile_shp = "TEST_POLYGON_crevasses_identified.shp"
outfile = "crevasse_factor_net.tif"

with rasterio.open(infile_raster) as src:
    profile = src.profile.copy()
    stability = src.read(1)
    transform = src.transform

gdf = gpd.read_file(infile_shp)
if gdf.crs != src.crs:
    gdf = gdf.to_crs(src.crs)

# Rasterise polygons: 1 inside polygons, 0 outside
mask = rasterize(
    [(geom, 1) for geom in gdf.geometry],
    out_shape=stability.shape,
    transform=transform,
    fill=0,
    dtype="uint8"
)

# Apply mask: set stability = 0 where polygons exist
stability_masked = stability.copy()
stability_masked[mask == 1] = 0 ############################################################# THIS IS THE VALUE GIVEN FOR CREVASSED REGIONS

if src.nodata is not None:
    stability_masked[stability == src.nodata] = src.nodata

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(stability_masked, 1)
```

## flow / buttressing factor

### calculating the MEDIAN KNMAX

```python
years = [2006, 2008, 2009, 2010, 2011, 2014, 2015, 2016, 2017, 2018]
arrays = []
meta = None

for year in years:
    filename = f"KNMAX_{year}_unmasked.tif"
    with rasterio.open(filename) as src:
        data = src.read(1).astype(float)
        data[data == src.nodata] = np.nan 
        arrays.append(data)
        if meta is None:
            meta = src.meta.copy()

stack = np.stack(arrays, axis=0)  
median_arr = np.nanmedian(stack, axis=0)   

meta.update(dtype="float32", count=1, nodata=np.nan)
with rasterio.open("KNMAX_MEDIAN.tif", "w", **meta) as dst:
    dst.write(median_arr.astype(np.float32), 1)
```

Then, in QGIS, I clipped the KNMAX_MEDIAN geotiff to the 2006 ice extent, to create the geotiff: KNMAX_MEDIAN_clipped.tif

### normalised LIMITED

```python
infile = "KNMAX_MEDIAN_clipped.tif"
outfile = "buttressing_LIMITED.tif"

with rasterio.open(infile) as src:
    knmax = src.read(1).astype(float)
    meta = src.meta.copy()

    if src.nodata is not None:
        knmax[knmax == src.nodata] = np.nan

    # Initialise normalised array
    norm_knmax = np.full_like(knmax, np.nan, dtype=float)

    # Case 1: KNMAX < 0  0
    norm_knmax[knmax < 0] = 0

    # Case 2: 0  KNMAX  1  scale linearly (same values)
    mask_mid = (knmax >= 0) & (knmax <= 1)
    norm_knmax[mask_mid] = knmax[mask_mid]  # already between 0 and 1

    # Case 3: KNMAX > 1  1
    norm_knmax[knmax > 1] = 1

    norm_knmax[np.isnan(knmax)] = np.nan

meta.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(outfile, "w", **meta) as dst:
    dst.write(norm_knmax.astype(np.float32), 1)
```

### normalised UNLIMITED

```python
infile = "KNMAX_MEDIAN_clipped.tif"
outfile = "buttressing_UNLIMITED.tif"

with rasterio.open(infile) as src:
    knmax = src.read(1).astype(float)
    meta = src.meta.copy()

    if src.nodata is not None:
        knmax[knmax == src.nodata] = np.nan

    # --- Normalisation (0 = weakest, 1 = strongest) ---
    valid = knmax[~np.isnan(knmax)]
    min_val, max_val = np.nanmin(valid), np.nanmax(valid)

    norm_knmax = (knmax - min_val) / (max_val - min_val)
    norm_knmax[np.isnan(knmax)] = np.nan

meta.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(outfile, "w", **meta) as dst:
    dst.write(norm_knmax.astype(np.float32), 1)
```

### PASSIVE POLYGON buttressing

```python
infile = "KNMAX_MEDIAN_clipped.tif"
shapefile = "passive_boundary_MEDIAN_polygons.shp"
outfile = "buttressing_PASSIVE.tif"

with rasterio.open(infile) as src:
    knmax = src.read(1).astype(float)
    meta = src.meta.copy()

    if src.nodata is not None:
        knmax[knmax == src.nodata] = np.nan

    norm_knmax = np.full_like(knmax, np.nan, dtype=float)
    # Case 1: KNMAX < 0  0
    norm_knmax[knmax < 0] = 0
    # Case 2: 0  KNMAX  1  scale linearly (same values)
    mask_mid = (knmax >= 0) & (knmax <= 1)
    norm_knmax[mask_mid] = knmax[mask_mid]  # already between 0 and 1
    # Case 3: KNMAX > 1  1
    norm_knmax[knmax > 1] = 1

    # Rasterise the passive polygon shapefile
    gdf = gpd.read_file(shapefile).to_crs(src.crs)  
    passive_mask = rasterize(
        [(geom, 1) for geom in gdf.geometry],
        out_shape=knmax.shape,
        transform=src.transform,
        fill=0,
        dtype="uint8"
    )

    norm_knmax[passive_mask == 1] = 0 # makes values within the shapefile boundary equal 0 #######################################################
    norm_knmax[np.isnan(knmax)] = np.nan

meta.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(outfile, "w", **meta) as dst:
    dst.write(norm_knmax.astype(np.float32), 1)
```

```python
outfile = "buttressing_PASSIVE.tif"
with rasterio.open(outfile) as src:
    knmax = src.read(1)  
    extent = (src.bounds.left,src.bounds.right,src.bounds.bottom,src.bounds.top,)  

plt.figure(figsize=(6, 4))
plt.imshow(knmax, cmap="Blues", extent=extent, origin="upper", vmin = 0, vmax = 1)
plt.colorbar(label="KNMAX (0 = passive, 1 = active)")
plt.title(" buttressing_PASSIVE ")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### Passive polygon PLUS FILL

```python

infile = "KNMAX_MEDIAN_clipped.tif"
passive_shapefile = "passive_boundary_MEDIAN_polygons.shp"
tensile075_shapefile = "tensile_stress_075.shp"
tensile100_shapefile = "tensile_stress_100_fill.shp"
outfile = "buttressing_PASSIVE_v2.tif"

with rasterio.open(infile) as src:
    knmax = src.read(1).astype(float)
    meta = src.meta.copy()

    if src.nodata is not None:
        knmax[knmax == src.nodata] = np.nan

    norm_knmax = np.full_like(knmax, np.nan, dtype=float)

    # Case 1: KNMAX < 0  0
    norm_knmax[knmax < 0] = 0
    # Case 2: 0  KNMAX  1  keep values
    mask_mid = (knmax >= 0) & (knmax <= 1)
    norm_knmax[mask_mid] = knmax[mask_mid]
    # Case 3: KNMAX > 1  1
    norm_knmax[knmax > 1] = 1

    # Passive polygons  fill with 0
    gdf_passive = gpd.read_file(passive_shapefile).to_crs(src.crs)
    passive_mask = rasterize(
        [(geom, 1) for geom in gdf_passive.geometry],
        out_shape=knmax.shape,
        transform=src.transform,
        fill=0,
        dtype="uint8"
    )
    norm_knmax[passive_mask == 1] = 0

    # Tensile stress 0.75 polygons  fill with 0.75
    gdf_tensile075 = gpd.read_file(tensile075_shapefile).to_crs(src.crs)
    tensile075_mask = rasterize(
        [(geom, 1) for geom in gdf_tensile075.geometry],
        out_shape=knmax.shape,
        transform=src.transform,
        fill=0,
        dtype="uint8"
    )
    norm_knmax[tensile075_mask == 1] = 0.75

    # Tensile stress 1 polygons  fill with 1
    gdf_tensile100 = gpd.read_file(tensile100_shapefile).to_crs(src.crs)
    tensile100_mask = rasterize(
        [(geom, 1) for geom in gdf_tensile100.geometry],
        out_shape=knmax.shape,
        transform=src.transform,
        fill=0,
        dtype="uint8"
    )
    norm_knmax[tensile100_mask == 1] = 1.0

    # Preserve NaNs where original raster was nodata
    norm_knmax[np.isnan(knmax)] = np.nan

meta.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(outfile, "w", **meta) as dst:
    dst.write(norm_knmax.astype(np.float32), 1)
```

## SMB factor

computing the median value across bands 56 to 104 (2006 to 2018)

```python
infile = "MEaSUREs_SMB_Quarterly_clipped.tif"
outfile = "MEaSUREs_SMB_Quarterly_clipped_MEDIAN.tif"

with rasterio.open(infile) as src:
    meta = src.meta.copy()

    arrays = []
    for b in range(56, 105):  
        band = src.read(b).astype(float)
        band[band == src.nodata] = np.nan 
        arrays.append(band)

    stack = np.stack(arrays, axis=0)  
    median_arr = np.nanmedian(stack, axis=0)

meta.update(dtype="float32", count=1, nodata=np.nan)

with rasterio.open(outfile, "w", **meta) as dst:
    dst.write(median_arr.astype(np.float32), 1)
```

```python
infile = "MEaSUREs_SMB_Quarterly_clipped_MEDIAN.tif"
outfile = "MEaSUREs_SMB.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    smb = src.read(1).astype(float)

    mask = np.isnan(smb)

    valid = smb[~mask]
    min_val, max_val = np.min(valid), np.max(valid)

    if min_val == max_val:
        norm = np.ones_like(smb, dtype=float)
    else:
        # Standard normalisation 0..1
        norm = (smb - min_val) / (max_val - min_val)
        # Invert so min  1, max  0
        norm = 1 - norm

    norm[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(norm.astype("float32"), 1)
```

```python
infile = "MEaSUREs_SMB_Quarterly_clipped_MEDIAN.tif"
outfile = "MEaSUREs_SMB_2.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    smb = src.read(1).astype(float)

    mask = np.isnan(smb)

    valid = smb[~mask]
    min_val = np.min(valid)

    # Define cutoff
    cutoff = 0.6

    # Initialize output array with 0s
    norm = np.zeros_like(smb, dtype=float)

    # Scale only values between min_val and cutoff
    scale_range = cutoff - min_val
    if scale_range > 0:
        # Linear mapping: min_val  1, cutoff  0
        scaled = (cutoff - smb) / scale_range
        norm = np.where((smb >= min_val) & (smb < cutoff), scaled, norm)

    # Apply mask
    norm[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(norm.astype("float32"), 1)
```

# COMBINED FRAMEWORK

```python
files = [
    "bmr_factor_WAOM_high.tif",
    "crevasse_factor_net.tif",
    "buttressing_PASSIVE_v2.tif",
    "MEaSUREs_SMB_2.tif"
]

for f in files:
    with rasterio.open(f) as src:
        print(f"{f}: resolution={src.res}, shape={src.shape}")
```

```python
files = [
    "bmr_factor_WAOM_high.tif",
    "crevasse_factor_net.tif",
    "buttressing_PASSIVE_v2.tif",
    "MEaSUREs_SMB_2.tif"
]

ref_file = files[1]  # use the second raster as reference
outfile = "total_framework.tif"

with rasterio.open(ref_file) as ref:
    ref_meta = ref.meta.copy()
    ref_arr = ref.read(1).astype(float)
    nodata = ref.nodata
    if nodata is not None:
        ref_arr[ref_arr == nodata] = np.nan

combined = np.ones_like(ref_arr, dtype=float)

for f in files:
    with rasterio.open(f) as src:
        arr = src.read(1).astype(float)
        if src.nodata is not None:
            arr[arr == src.nodata] = np.nan

        # Reproject/resample to match reference grid
        reprojected = np.empty_like(ref_arr, dtype=float)
        reproject(
            source=arr,
            destination=reprojected,
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=ref_meta['transform'],
            dst_crs=ref_meta['crs'],
            resampling=Resampling.bilinear
        )

        # Mask nodata as NaN
        reprojected[reprojected == src.nodata] = np.nan

        # Multiply in
        combined *= reprojected

# Mask out NaNs
combined[np.isnan(ref_arr)] = np.nan

# Save
ref_meta.update(dtype="float32", count=1, nodata=np.nan)
with rasterio.open(outfile, "w", **ref_meta) as dst:
    dst.write(combined.astype("float32"), 1)
```

```python
outfile = "total_framework.tif"
with rasterio.open(outfile) as src:
    total = src.read(1)  
    extent = (src.bounds.left,src.bounds.right,src.bounds.bottom,src.bounds.top,)  

plt.figure(figsize=(6, 4))
plt.imshow(total, cmap="Reds", extent=extent, origin="upper", vmin = 0, vmax = 1)
plt.colorbar(label="Combined Framework (0 = Least Ideal, 1 = Most Ideal)")
plt.title(" total_framework ")
plt.xlabel("Eastings (m)")
plt.ylabel("Northings (m)")
plt.show()
```

Normalising this combined framework

```python
infile = "total_framework.tif"
outfile = "total_framework_normalised.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    data = src.read(1).astype(float)

    mask = np.isnan(data)
    if src.nodata is not None:
        data[data == src.nodata] = np.nan
        mask = np.isnan(data)

    # Compute min and max (ignoring NaNs)
    valid = data[~mask]
    min_val, max_val = np.min(valid), np.max(valid)

    if min_val == max_val:
        # Avoid divide-by-zero if raster is constant
        norm = np.zeros_like(data, dtype=float)
    else:
        norm = (data - min_val) / (max_val - min_val)

    # Restore NaNs
    norm[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(norm.astype(np.float32), 1)
```

```python
outfile = "total_framework_normalised.tif"
with rasterio.open(outfile) as src:
    total = src.read(1)  
    extent = (src.bounds.left,src.bounds.right,src.bounds.bottom,src.bounds.top,)  

plt.figure(figsize=(6, 4))
plt.imshow(total, cmap="Reds", extent=extent, origin="upper", vmin = 0, vmax = 1)
plt.colorbar(label="Comb. Framework (0 = Least Ideal, 1 = Most Ideal)")
plt.title(" total_framework_normalised ")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

### Calculating the area of pixels greater than 0

```python
raster_path = "total_framework_normalised.tif"

with rasterio.open(raster_path) as src:
    data = src.read(1)
    transform = src.transform
    crs = src.crs
    
    print(f"CRS: {crs}") 
    print(f"Pixel size: {transform[0]} x {abs(transform[4])} meters")

    # Masks
    mask_gt0 = data > 0        # values > 0
    mask_ge05 = data >= 0.5    # values >= 0.5
    mask_ge08 = data >= 0.8    # values >= 0.8

    # Count pixels
    num_pixels_gt0 = np.sum(mask_gt0)
    num_pixels_ge05 = np.sum(mask_ge05)
    num_pixels_ge08 = np.sum(mask_ge08)

    # Pixel area (in square meters)
    pixel_width = transform[0]
    pixel_height = -transform[4]  # usually negative in affine transform
    pixel_area = abs(pixel_width * pixel_height)

    # Total areas
    total_area_gt0_m2 = num_pixels_gt0 * pixel_area
    total_area_gt0_km2 = total_area_gt0_m2 / 1e6

    total_area_ge05_m2 = num_pixels_ge05 * pixel_area
    total_area_ge05_km2 = total_area_ge05_m2 / 1e6

    total_area_ge08_m2 = num_pixels_ge08 * pixel_area
    total_area_ge08_km2 = total_area_ge08_m2 / 1e6

    # Results
    print("=== Area Calculations ===")
    print(f"Pixels > 0: {num_pixels_gt0}")
    print(f"  Total area = {total_area_gt0_m2:.2f} m ({total_area_gt0_km2:.2f} km)\n")

    print(f"Pixels >= 0.5: {num_pixels_ge05}")
    print(f"  Total area = {total_area_ge05_m2:.2f} m ({total_area_ge05_km2:.2f} km)\n")

    print(f"Pixels >= 0.8: {num_pixels_ge08}")
    print(f"  Total area = {total_area_ge08_m2:.2f} m ({total_area_ge08_km2:.2f} km)")
```

# Weighted Framework

```python
files = [
    "bmr_factor_WAOM_high.tif",
    "crevasse_factor_net.tif",
    "buttressing_PASSIVE_v2.tif",
    "MEaSUREs_SMB_2.tif"
]

for f in files:
    with rasterio.open(f) as src:
        print(f"{f}: resolution={src.res}, shape={src.shape}")
```

```python
import numpy as np
import rasterio
from rasterio.warp import reproject, Resampling

files = {
    "melt": "bmr_factor_WAOM_high.tif",
    "crevasse": "crevasse_factor_net.tif",
    "buttress": "buttressing_PASSIVE_v2.tif",
    "smb": "MEaSUREs_SMB_2.tif"
}

outfile = "total_framework_WEIGHTED.tif"

# Weights
wm, wb = 0.7, 0.3   # melt and buttressing
wc, ws = 1, 0.8   # multipliers for crevassing and SMB (suitability)

# Use crevasse raster as reference grid
ref_file = files["crevasse"]
with rasterio.open(ref_file) as ref:
    ref_meta = ref.meta.copy()
    ref_arr = ref.read(1).astype(float)
    nodata = ref.nodata
    if nodata is not None:
        ref_arr[ref_arr == nodata] = np.nan

# Function to load and reproject rasters onto reference grid
def load_reproject(file, ref_meta, ref_arr):
    with rasterio.open(file) as src:
        arr = src.read(1).astype(float)
        if src.nodata is not None:
            arr[arr == src.nodata] = np.nan
        reprojected = np.empty_like(ref_arr, dtype=float)
        reproject(
            source=arr,
            destination=reprojected,
            src_transform=src.transform,
            src_crs=src.crs,
            dst_transform=ref_meta['transform'],
            dst_crs=ref_meta['crs'],
            resampling=Resampling.bilinear
        )
        return reprojected

# Load each factor
melt_arr = load_reproject(files["melt"], ref_meta, ref_arr)
buttress_arr = load_reproject(files["buttress"], ref_meta, ref_arr)
smb_arr = load_reproject(files["smb"], ref_meta, ref_arr)
crev_arr = load_reproject(files["crevasse"], ref_meta, ref_arr)

# Compute weighted additive part
additive = wm * melt_arr + wb * buttress_arr

# Multiply by suitability factors (higher = better)
score = additive * (wc * crev_arr) * (ws * smb_arr)

# Mask out reference NaNs
score[np.isnan(ref_arr)] = np.nan

# Save output
ref_meta.update(dtype="float32", count=1, nodata=np.nan)
with rasterio.open(outfile, "w", **ref_meta) as dst:
    dst.write(score.astype("float32"), 1)
```

normalising this combined framework

```python
infile = "total_framework_WEIGHTED.tif"
outfile = "total_framework_WEIGHTED_normalised.tif"

with rasterio.open(infile) as src:
    profile = src.profile.copy()
    data = src.read(1).astype(float)

    mask = np.isnan(data)
    if src.nodata is not None:
        data[data == src.nodata] = np.nan
        mask = np.isnan(data)

    # Compute min and max (ignoring NaNs)
    valid = data[~mask]
    min_val, max_val = np.min(valid), np.max(valid)

    if min_val == max_val:
        # Avoid divide-by-zero if raster is constant
        norm = np.zeros_like(data, dtype=float)
    else:
        norm = (data - min_val) / (max_val - min_val)

    # Restore NaNs
    norm[mask] = np.nan

profile.update(dtype="float32", nodata=np.nan)

with rasterio.open(outfile, "w", **profile) as dst:
    dst.write(norm.astype(np.float32), 1)
```
