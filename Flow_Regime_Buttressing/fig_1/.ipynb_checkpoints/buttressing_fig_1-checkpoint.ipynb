```python
from matplotlib.colors import ListedColormap
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd 
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
import matplotlib.cm as cm
import matplotlib.colors as mcolors
```

```python
xlimit = (1025000, 1140000)
ylimit = (-2148431.9935, -2033743.9935)

xlimit2 = (1060036.7411, 1119656.7298)
ylimit2 = (-2135000, -2075000)
```

```python
tif_path = "knmax_MEAN_MASKED.tif"
with rasterio.open(tif_path) as src:
    knmax_mean = src.read(1)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2006.shp')
passive_boundary = gpd.read_file('passive_boundary.shp')

with rasterio.open("knmax_MEAN_MASKED_CLIPPED.tif") as src:
    second_image = src.read(1).astype(float)  # Convert to float so it can hold NaN
    second_extent = [
        src.bounds.left, src.bounds.right,
        src.bounds.bottom, src.bounds.top
    ]

second_image[second_image == 0] = np.nan
```

```python
fig, axes = plt.subplots(1, 2, figsize=(160/25.4, 160/25.4))

######################################################### SUBPLOT 1
ax1 = axes[0]
masked_knmax_mean = np.ma.masked_invalid(knmax_mean)

img1 = ax1.imshow(masked_knmax_mean, cmap='RdBu', origin='upper', extent=extent, vmin=-5, vmax=5)

groundingline.plot(ax=ax1, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax1, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider1 = make_axes_locatable(ax1)
cax1 = divider1.append_axes("right", size="5%", pad=0.1)
cbar1 = fig.colorbar(img1, cax=cax1)
ticks = np.linspace(-5, 5, 5)  
cbar1.set_ticks(ticks)
for label in cbar1.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(10)
cbar1.set_label("KNMAX", fontsize=10, fontfamily='Times New Roman')

ax1.set_xlabel('Eastings (m)', fontsize=10, fontfamily='Times New Roman')
ax1.set_ylabel('Northings (m)', fontsize=10, fontfamily='Times New Roman')
for label in (ax1.get_xticklabels() + ax1.get_yticklabels()):
    label.set_fontsize(10)
    label.set_family('Times New Roman')
ax1.xaxis.get_offset_text().set_fontname('Times New Roman')
ax1.xaxis.get_offset_text().set_fontsize(10)
ax1.yaxis.get_offset_text().set_fontname('Times New Roman')
ax1.yaxis.get_offset_text().set_fontsize(10)
ax1.tick_params(direction='in')
ax1.set_xlim(xlimit)
ax1.set_ylim(ylimit)
ax1.set_aspect('equal')
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.35), ncol=2, prop={'size': 8, 'family': 'Times New Roman'}, title_fontsize=8)


######################################################### SUBPLOT 2
ax2 = axes[1]
masked_knmax_mean = np.ma.masked_invalid(knmax_mean)
solid_cmap = ListedColormap(['white'])

#img2 = ax2.imshow(masked_knmax_mean, cmap=solid_cmap, origin='upper', extent=extent, vmin=0, vmax=1)
second_img_plot = ax2.imshow(second_image, cmap='Greys', extent=second_extent, origin='upper', vmin=-5, vmax=5)

groundingline.plot(ax=ax2, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax2, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

# Passive ice boundary colours
years = passive_boundary['Year'].unique()
years.sort()
n = len(years) + 2
cmap = cm.get_cmap('plasma', n)
norm = mcolors.Normalize(vmin=years.min(), vmax=years.max())
color_map = {year: cmap(i / (n - 1)) for i, year in enumerate(years)}

legend_handles = []
for year in years:
    color = color_map[year]
    subset = passive_boundary[passive_boundary['Year'] == year]
    for geom in subset.geometry:
        if geom.geom_type == 'MultiLineString':
            for part in geom.geoms:
                xs, ys = part.xy
                ax2.plot(xs, ys, color=color, linewidth=1, label=None)
        elif geom.geom_type == 'LineString':
            xs, ys = geom.xy
            ax2.plot(xs, ys, color=color, linewidth=1, label=None)
    patch = mpatches.Patch(color=color, label=str(year))
    legend_handles.append(patch)

legend = ax2.legend(
    handles=legend_handles,
    title='Passive Ice Boundaries',
    loc='upper center',                
    bbox_to_anchor=(0.5, -0.25),       
    borderaxespad=0,
    ncol=4,                            
    prop={'size': 8, 'family': 'Times New Roman'},
    title_fontsize=8
)
legend.get_frame().set_linewidth(1)
legend.get_title().set_fontsize(9)
legend.get_title().set_fontfamily('Times New Roman')

divider2 = make_axes_locatable(ax2)
cax2 = divider2.append_axes("right", size="5%", pad=0.1)
cbar2 = fig.colorbar(second_img_plot, cax=cax2)
cbar2.set_ticks(ticks)
for label in cbar2.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(10)
cbar2.set_label("KNMAX", fontsize=10, fontfamily='Times New Roman')

ax2.set_xlabel('Eastings (m)', fontsize=10, fontfamily='Times New Roman')
ax2.set_ylabel('Northings (m)', fontsize=10, fontfamily='Times New Roman')
for label in (ax2.get_xticklabels() + ax2.get_yticklabels()):
    label.set_fontsize(10)
    label.set_family('Times New Roman')
ax2.xaxis.get_offset_text().set_fontname('Times New Roman')
ax2.xaxis.get_offset_text().set_fontsize(10)
ax2.yaxis.get_offset_text().set_fontname('Times New Roman')
ax2.yaxis.get_offset_text().set_fontsize(10)
ax2.tick_params(direction='in')
ax2.set_xlim(xlimit2)
ax2.set_ylim(ylimit2)
ax2.set_aspect('equal')

######################################################### 
ax1.text(0.05, 0.95, "a.", transform=ax1.transAxes, fontsize=10, fontname = 'Times New Roman', fontweight='bold', va='top', ha='left')
ax2.text(0.05, 0.95, "b.", transform=ax2.transAxes, fontsize=10, fontname = 'Times New Roman', fontweight='bold', va='top', ha='left')

plt.tight_layout(pad=0.3, w_pad=0.3)
#plt.savefig('figure_1.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

```python
from matplotlib.colors import ListedColormap
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd 
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
import matplotlib.cm as cm
import matplotlib.colors as mcolors
```

```python
xlimit = (998648.2458, 1171935.9538)
ylimit = (-2148431.9935, -2033743.9935)

tif_path = "knmax_MEAN_MASKED.tif"
with rasterio.open(tif_path) as src:
    knmax_mean = src.read(1)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2006.shp')
passive_boundary = gpd.read_file('passive_boundary.shp')


with rasterio.open("knmax_MEAN_MASKED_CLIPPED.tif") as src:
    second_image = src.read(1).astype(float)  # Convert to float so it can hold NaN
    second_extent = [
        src.bounds.left, src.bounds.right,
        src.bounds.bottom, src.bounds.top
    ]

second_image[second_image == 0] = np.nan
```

```python
xlimit2 = (1060036.7411, 1119656.7298)
ylimit2 = (-2131256.0030, -2072866.4487)


fig, ax = plt.subplots(figsize=(160/25.4, 160/25.4))  

masked_data = np.ma.masked_invalid(knmax_mean)
solid_cmap = ListedColormap(['white'])

# base layers:
img = ax.imshow(masked_data, cmap=solid_cmap, origin='upper', extent=extent, vmin=0, vmax=1)
second_img_plot = ax.imshow(second_image,cmap='Greys',extent=second_extent,origin='upper',vmin=-5, vmax=5)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

# passive ice boundary colours
years = passive_boundary['Year'].unique()
years.sort()
n = len(years) + 2
cmap = cm.get_cmap('plasma', n)
norm = mcolors.Normalize(vmin=years.min(), vmax=years.max())
color_map = {year: cmap(i / (n - 1)) for i, year in enumerate(years)}

legend_handles = []
for year in years:
    color = color_map[year]
    subset = passive_boundary[passive_boundary['Year'] == year]

    for geom in subset.geometry:
        if geom.geom_type == 'MultiLineString':
            for part in geom.geoms:
                xs, ys = part.xy
                ax.plot(xs, ys, color=color, linewidth=2, label=None)
        elif geom.geom_type == 'LineString':
            xs, ys = geom.xy
            ax.plot(xs, ys, color=color, linewidth=2, label=None)

    patch = mpatches.Patch(color=color, label=str(year))
    legend_handles.append(patch)

legend = ax.legend(handles=legend_handles,
    title='Passive Boundaries',
    loc='upper center',                
    bbox_to_anchor=(0.5, -0.1),       
    borderaxespad=0,
    ncol=4,                            
    prop={'size': 9, 'family': 'Times New Roman'},
    title_fontsize=10
)
legend.get_frame().set_linewidth(1)
legend.get_title().set_fontsize(10)
legend.get_title().set_fontfamily('Times New Roman')


divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(second_img_plot, cax=cax)
ticks = np.linspace(-5, 5, 5)  
cbar.set_ticks(ticks)
for label in cbar.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(10)
cbar.set_label("KNMAX", fontsize=10, fontfamily='Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=10, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=10, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(10)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(10)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(10)
ax.tick_params(direction='in')
ax.set_xlim(xlimit2)
ax.set_ylim(ylimit2)
# ax.set_title("Mean KNMAX with passive ice boundary evolution", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

# legend2 = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
# legend2.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('passive_ice_grey.jpg', dpi=500, bbox_inches='tight')
plt.show()
```
