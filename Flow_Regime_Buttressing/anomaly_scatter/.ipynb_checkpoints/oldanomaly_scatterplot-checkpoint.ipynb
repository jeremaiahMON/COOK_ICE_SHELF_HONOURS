```python
# For Thickness Data:

# EXTRACTS THE DESIGNATED BAND FOR THE SPECIIC YEAR 

Fig2AThicknessResults = [("Band", "Mean Anomaly", "Median Anomaly")]

for band_number in range(1, 105):
    raster_path = "Fig2QuarterlyThickness.tif"
    with rasterio.open(raster_path) as src:
        band_data = src.read(band_number) 
        
    out_raster = f"Fig2Aband{band_number}.tif"
    with rasterio.open(raster_path) as src:
        profile = src.profile  
        profile.update(count=1) 
    
        with rasterio.open(out_raster, "w", **profile) as dst:
            dst.write(band_data, 1)
    
    ##########################################
    
    # CALCULATES THE ANOMALY BETWEEN THE SPECIFIC YEAR BAND AND THE MEAN BAND
    
    raster1_path = f"Fig2Aband{band_number}.tif"  
    raster2_path = "Fig2MeanThickness.tif" 
    output_path = f"Fig2Aanom{band_number}.tif"
    
    with rasterio.open(raster1_path) as src1, rasterio.open(raster2_path) as src2:
        raster1 = src1.read(1) 
        raster2 = src2.read(1)  
    
        difference = raster1 - raster2
    
        difference[difference == 0] = np.nan 
    
        profile = src1.profile
        profile.update(dtype=rasterio.float32, count=1, nodata=np.nan)
    
        with rasterio.open(output_path, "w", **profile) as dst:
            dst.write(difference, 1)
        
    ##########################################
    
    # CALCULATE THE MEAN AND MEDIAN OF THE ANOMALY ACROSS THE ENTIRE ICE SHEET
    
    raster3_path = f"Fig2Aanom{band_number}.tif"
    
    with rasterio.open(raster3_path) as src:
        
        raster3_data = src.read(1)
        
        nodata_value = src.nodata
    
        if nodata_value is not None:
            raster3_data = np.where(raster3_data == nodata_value, np.nan, raster3_data)
    
        mean_value = np.nanmean(raster3_data)
        median_value = np.nanmedian(raster3_data)

    Fig2AThicknessResults.append((band_number, mean_value, median_value))

output_csv = "Fig1AThicknessResults.csv"

with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(Fig2AThicknessResults)
```
