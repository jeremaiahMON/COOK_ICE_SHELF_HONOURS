### Calculating the Thickness Anomaly

```python
import rasterio
import numpy as np
import csv
```

```python
raster_path = 'Thickness_Quarterly.tif'
Thickness_Anomaly_Results = [("Band", "Mean Anomaly", "Median Anomaly")]

with rasterio.open(raster_path) as src:
    nodata = src.nodata
    height, width = src.height, src.width

    data_stack = []

    for band_number in range(56, 105):  # Bands 56 to 104 represent the years 2006 to 2018. See 'Thickness_BMR_Time_Stamps.csv' for more info.
        band_data = src.read(band_number).astype(np.float32)
        if nodata is not None:
            band_data = np.where(band_data == nodata, np.nan, band_data)

        data_stack.append(band_data)

    data_stack = np.stack(data_stack, axis=0)
    mean_thickness = np.nanmean(data_stack, axis=0) # This calculates the mean value across all data from 2006 to 2018.

with rasterio.open(raster_path) as src:
    for band_number in range(56, 105):
        band_data = src.read(band_number).astype(np.float32)

        if nodata is not None:
            band_data = np.where(band_data == nodata, np.nan, band_data)

        anomaly = band_data - mean_thickness # This calculates the anomaly value by subtracting the mean from a specific quarterly value.
        anomaly[anomaly == 0] = np.nan  

        mean_value = np.nanmean(anomaly)
        median_value = np.nanmedian(anomaly)

        Thickness_Anomaly_Results.append((band_number, mean_value, median_value))

# Saving Results to a CSV file:
output_csv = "Thickness_Anomaly_Results.csv"
with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(Thickness_Anomaly_Results)
```

### Calculating the BMR Anomaly

```python
import rasterio
import numpy as np
import csv
```

```python
raster_path = 'BMR_Quarterly.tif'
BMR_Anomaly_Results = [("Band", "Mean Anomaly", "Median Anomaly")]

with rasterio.open(raster_path) as src:
    nodata = src.nodata
    height, width = src.height, src.width

    data_stack = []

    for band_number in range(56, 105):  # Bands 56 to 104 represent the years 2006 to 2018. See 'Thickness_BMR_Time_Stamps.csv' for more info.
        band_data = src.read(band_number).astype(np.float32)
        if nodata is not None:
            band_data = np.where(band_data == nodata, np.nan, band_data)

        data_stack.append(band_data)

    data_stack = np.stack(data_stack, axis=0)
    mean_BMR = np.nanmean(data_stack, axis=0) # This calculates the mean value across all data from 2006 to 2018.

with rasterio.open(raster_path) as src:
    for band_number in range(56, 105):
        band_data = src.read(band_number).astype(np.float32)

        if nodata is not None:
            band_data = np.where(band_data == nodata, np.nan, band_data)

        anomaly = band_data - mean_BMR # This calculates the anomaly value by subtracting the mean from a specific quarterly value.
        anomaly[anomaly == 0] = np.nan  

        mean_value = np.nanmean(anomaly)
        median_value = np.nanmedian(anomaly)

        BMR_Anomaly_Results.append((band_number, mean_value, median_value))

# Saving Results to a CSV file:
output_csv = "BMR_Anomaly_Results.csv"
with open(output_csv, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(BMR_Anomaly_Results)
```

### Formatting the CSV files with TIME to prepare them for plotting

```python
import pandas as pd
```

```python
Time_Stamps = pd.read_csv('Thickness_BMR_Time_Stamps.csv')

Thickness_Time = pd.read_csv('Thickness_Anomaly_Results.csv')
Thickness_Time['Year'] = (Time_Stamps['BandInfo:DaysSince1/1/1950'] / 365) + 1950
Thickness_Time.to_csv('Thickness_Time.csv', index=False)
Thickness_Time = pd.read_csv('Thickness_Time.csv')

BMR_Time = pd.read_csv('BMR_Anomaly_Results.csv')
BMR_Time['Year'] = (Time_Stamps['BandInfo:DaysSince1/1/1950'] / 365) + 1950
BMR_Time.to_csv('BMR_Time.csv', index=False)
BMR_Time = pd.read_csv('BMR_Time.csv')
```

### Plotting Data

```python
import numpy as np 
from matplotlib import pyplot as plt 
from matplotlib import rcParams
```

```python
x = BMR_Time['Mean Anomaly']
y = Thickness_Time['Mean Anomaly']
```

```python
rcParams['font.family'] = 'Times New Roman'
rcParams['font.size'] = 12
plt.figure().set_size_inches(120/25.4, 100/25.4,forward=True)
plt.scatter(x,y,label='Mean Thickness Anomaly', marker='x', linewidths=0.5, c='m')
plt.xlabel('BMR Anomaly (m/yr)')
plt.ylabel('Thickness Anomaly (m)')
plt.tick_params(direction='in')
plt.axis('equal')

plt.tight_layout()
#plt.savefig('BMR_Thickness_Anomaly_Scatter.png', dpi=500, bbox_inches='tight')
plt.show()
```
