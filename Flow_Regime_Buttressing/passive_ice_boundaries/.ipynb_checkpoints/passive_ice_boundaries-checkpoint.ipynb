 - knmax data from matlab codes
 - hand drawn passive ice boundaries where knmax ~ 0
 - data sources: https://docs.google.com/document/d/1SUdnGJrY-3BL0muc6Ji8TW8O6DwPuUbxnuRvm83jew0/edit?tab=t.0
 - grounding line is 2003, ice front is 2006

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd 
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
xlimit = (998648.2458, 1171935.9538)
ylimit = (-2148431.9935, -2033743.9935)

tif_path = "knmax_MEAN_MASKED.tif"
with rasterio.open(tif_path) as src:
    knmax_mean = src.read(1)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2006.shp')
passive_boundary = gpd.read_file('passive_boundary.shp')
```

### normal extent, same colour

```python
fig, ax = plt.subplots(figsize=(160/25.4, 100/25.4))  # mm to inches

img = ax.imshow(knmax_mean, cmap='RdBu', origin='upper', extent=extent, vmin=-5, vmax = 5)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')
passive_boundary.plot(ax=ax, edgecolor='red', facecolor='none', linewidth=0.5, label='Passive Ice Boundary')

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
ticks = np.linspace(-5, 5, 5)  
cbar.set_ticks(ticks)
for label in cbar.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(12)
cbar.set_label("Depth (m)", fontsize=12, fontfamily='Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('passive_ice_1.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

### zoomed in extent, different colour. but still hard to see

```python
xlimit2 = (1029036.7411, 1119656.7298)
ylimit2 = (-2131256.0030, -2072866.4487)


fig, ax = plt.subplots(figsize=(160/25.4, 100/25.4))  # mm to inches

img = ax.imshow(knmax_mean, cmap='RdBu', origin='upper', extent=extent, vmin=-5, vmax = 5)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')
#passive_boundary.plot(ax=ax, edgecolor='red', facecolor='none', linewidth=0.5, label='Passive Ice Boundary')


import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Choose a colormap and create a list of colors
n = len(passive_boundary)
cmap = cm.get_cmap('Greens', n)  # use any colormap you like
colors = [cmap(i) for i in range(n)]

# Plot each geometry with a unique color
for i, (idx, row) in enumerate(passive_boundary.iterrows()):
    geom = row.geometry
    color = colors[i]

    if geom.geom_type == 'MultiLineString':
        for part in geom.geoms:
            xs, ys = part.xy
            ax.plot(xs, ys, color=color, linewidth=0.8, label=None)
    elif geom.geom_type == 'LineString':
        xs, ys = geom.xy
        ax.plot(xs, ys, color=color, linewidth=0.8, label=None)




divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
ticks = np.linspace(-5, 5, 5)  
cbar.set_ticks(ticks)
for label in cbar.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(12)
cbar.set_label("Depth (m)", fontsize=12, fontfamily='Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit2)
ax.set_ylim(ylimit2)
ax.set_title("Mean KNMAX with passive ice boundary evolution", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('passive_ice_2.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

### zoomed in, larger

```python
xlimit2 = (1060036.7411, 1119656.7298)
ylimit2 = (-2131256.0030, -2072866.4487)


fig, ax = plt.subplots(figsize=(12, 12))  

img = ax.imshow(knmax_mean, cmap='RdBu', origin='upper', extent=extent, vmin=-5, vmax = 5)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')
#passive_boundary.plot(ax=ax, edgecolor='red', facecolor='none', linewidth=0.5, label='Passive Ice Boundary')



#######################################################################################

import matplotlib.patches as mpatches
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Setup colormap
years = passive_boundary['Year'].unique()
years.sort()
n = len(years)
cmap = cm.get_cmap('plasma', n)
norm = mcolors.Normalize(vmin=years.min(), vmax=years.max())
color_map = {year: cmap(i / (n - 1)) for i, year in enumerate(years)}

# Store legend handles
legend_handles = []

# Plot each year's geometry in its own color
for year in years:
    color = color_map[year]
    subset = passive_boundary[passive_boundary['Year'] == year]

    for geom in subset.geometry:
        if geom.geom_type == 'MultiLineString':
            for part in geom.geoms:
                xs, ys = part.xy
                ax.plot(xs, ys, color=color, linewidth=2, label=None)
        elif geom.geom_type == 'LineString':
            xs, ys = geom.xy
            ax.plot(xs, ys, color=color, linewidth=2, label=None)

    # Create legend entry for this year
    patch = mpatches.Patch(color=color, label=str(year))
    legend_handles.append(patch)

# Add the legend to your figure
# ax.legend(handles=legend_handles, title='Passive Ice Boundaries by Year',
#           loc='lower right', prop={'size': 9, 'family': 'Times New Roman'},
#           title_fontsize=10).get_frame().set_linewidth(0)

# Create and position the legend outside the plot (right side)
legend = ax.legend(
    handles=legend_handles,
    title='Passive Boundaries',
    loc='center left',
    bbox_to_anchor=(1.02, 0.5),
    borderaxespad=5,
    prop={'size': 9, 'family': 'Times New Roman'},
    title_fontsize=10
)
legend.get_frame().set_linewidth(0)

#######################################################################################




divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
ticks = np.linspace(-5, 5, 5)  
cbar.set_ticks(ticks)
for label in cbar.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(12)
cbar.set_label("Depth (m)", fontsize=12, fontfamily='Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit2)
ax.set_ylim(ylimit2)
ax.set_title("Mean KNMAX with passive ice boundary evolution", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

#legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
#legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('passive_ice_3.jpg', dpi=500, bbox_inches='tight')
plt.show()
```
