### Subplot b: WSB bed elevation - from bedmap 2 quantartica

```python
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
```

```python
tif_path = r'C:\Users\jjls0\OneDrive\Documents\HONOURS\PROGRAMS\zzResults_Figures\Other\Spatial_context\Bed_Elevation_BedMap2_clipped.tif'
with rasterio.open(tif_path) as src:
    bed = src.read(1)
    extent1 = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

groundingline1 = gpd.read_file(r'C:\Users\jjls0\OneDrive\Documents\HONOURS\PROGRAMS\zzResults_Figures\Other\Spatial_context\2003_GL_full.shp')
```

```python
fig, ax = plt.subplots(figsize=(10,8))  
img = ax.imshow(bed, cmap='terrain', origin='upper', extent=extent)

groundingline1.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)


divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
cbar.set_label("Bed Elevation (m)", fontsize=12, fontfamily='Times New Roman')
for tick_label in cbar.ax.get_yticklabels():
    tick_label.set_fontsize(12)
    tick_label.set_family('Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')

ax.xaxis.get_offset_text().set_fontsize(12)
ax.xaxis.get_offset_text().set_family('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_family('Times New Roman')

ax.tick_params(direction='in')
ax.set_xlim(xlimit1)
ax.set_ylim(ylimit1)
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_edgecolor('none')  # Removes the legend border

for spine in ax.spines.values():
    spine.set_edgecolor('red')
    spine.set_linewidth(2)  

plt.tight_layout()
#plt.savefig('WAOM_Cross_Section_Cords.jpg', dpi=500, bbox_inches='tight')
plt.show()
```
