## BMR with no smoothing

```python
import rasterio
import matplotlib.pyplot as plt
import geopandas as gpd 
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
xlimit = (998648.2458, 1171935.9538)
ylimit = (-2148431.9935, -2033743.9935)

tif_path = "CIS_BMR_mask_clipped.tif"

with rasterio.open(tif_path) as src:
    melt = src.read(1)  # Read the first band
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

s_per_yr = 365.25 * 24 * 60 * 60 
melt = melt * s_per_yr  # convert to m/yr

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2011.shp')
```

```python
# Plotting the data

fig, ax = plt.subplots(figsize=(160/25.4, 100/25.4))  # mm to inches
img = ax.imshow(melt, cmap='viridis', origin='upper', extent=extent)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
cbar.set_label("BMR (m/yr)", fontsize=12, fontfamily='Times New Roman')
for tick in cbar.ax.get_yticklabels():
    tick.set_fontname('Times New Roman')
    tick.set_fontsize(12)

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)

ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
#ax.set_title("BMR (WAOM)", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('WAOM_BMR.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

## BMR with Gaussian smoothing

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from mpl_toolkits.axes_grid1 import make_axes_locatable
import geopandas as gpd 
```

```python
# Prepping data for plotting:

xlimit = (998648.2458, 1171935.9538)
ylimit = (-2148431.9935, -2033743.9935)

tif_path = "CIS_BMR_mask_clipped.tif"
with rasterio.open(tif_path) as src:
    melt = src.read(1)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

# Convert from m/s to m/yr
s_per_yr = 365.25 * 24 * 60 * 60 
melt = melt * s_per_yr

# Applying the Gaussian smoothing
sigma = 1 
melt_smooth = gaussian_filter(np.nan_to_num(melt, nan=0.0), sigma=sigma)
melt_smooth[np.isnan(melt)] = np.nan

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2011.shp')
```

```python
# Plotting the data

fig, ax = plt.subplots(figsize=(160/25.4, 100/25.4))  # mm to inches
img = ax.imshow(melt_smooth, cmap='viridis', origin='upper', extent=extent)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
cbar.set_label("BMR (m/yr)", fontsize=12, fontfamily='Times New Roman')
for tick in cbar.ax.get_yticklabels():
    tick.set_fontname('Times New Roman')
    tick.set_fontsize(12)

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
#ax.set_title("Gaussian Smoothed BMR (WAOM)", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('WAOM_BMR_Smoothed.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

# Ice Draft Depth Visualisation

Indicative of pressure-induced melt

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
from mpl_toolkits.axes_grid1 import make_axes_locatable
import geopandas as gpd 
```

```python
# Prepping data for plotting:

xlimit = (998648.2458, 1171935.9538)
ylimit = (-2148431.9935, -2033743.9935)

tif_path = "zice_3031_clipped.tif"
with rasterio.open(tif_path) as src:
    depth = src.read(1)
    extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]

# Applying the Gaussian smoothing
sigma = 1 
depth_smooth = gaussian_filter(np.nan_to_num(depth, nan=0.0), sigma=sigma)
depth_smooth[np.isnan(depth)] = np.nan

groundingline = gpd.read_file('2003_GL_full_cook_complete_dissolved_larger.shp')
coastline = gpd.read_file('ice_front_2011.shp')
```

```python
# Plotting the data

fig, ax = plt.subplots(figsize=(160/25.4, 100/25.4))  # mm to inches

img = ax.imshow(depth_smooth, cmap='Blues_r', origin='upper', extent=extent, vmin=-1000, vmax = 0)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
cbar = fig.colorbar(img, cax=cax)
ticks = np.linspace(-1000, 0, 6)  
cbar.set_ticks(ticks)
cbar.set_ticklabels([f"{int(abs(t))}" for t in ticks])  
for label in cbar.ax.get_yticklabels():
    label.set_fontname('Times New Roman')
    label.set_fontsize(12)
cbar.set_label("Depth (m)", fontsize=12, fontfamily='Times New Roman')

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
#ax.set_title("Gaussian Smoothed Ice Draft Depth (WAOM)", fontsize=12, fontfamily='Times New Roman')
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

plt.tight_layout()
#plt.savefig('WAOM_Ice_Draft_smoothed.jpg', dpi=500, bbox_inches='tight')
plt.show()
```

# Plotting BMR and Ice Draft Together

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
```

```python
fig, axes = plt.subplots(2, 1, figsize=(160/25.4, 200/25.4)) 

# BMR:
ax = axes[0]
img1 = ax.imshow(melt_smooth, cmap='viridis', origin='upper', extent=extent)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider1 = make_axes_locatable(ax)
cax1 = divider1.append_axes("right", size="5%", pad=0.1)
cbar1 = fig.colorbar(img1, cax=cax1)
cbar1.set_label("BMR (m/yr)", fontsize=12, fontfamily='Times New Roman')
for tick in cbar1.ax.get_yticklabels():
    tick.set_fontname('Times New Roman')
    tick.set_fontsize(12)

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

# Depth:
ax = axes[1]
img2 = ax.imshow(depth_smooth, cmap='Blues_r', origin='upper', extent=extent, vmin=-1000, vmax=0)

groundingline.plot(ax=ax, edgecolor='black', facecolor='azure', linewidth=2, label='Grounding Line', alpha=0.8)
coastline.plot(ax=ax, edgecolor='green', facecolor='none', linewidth=2, label='Ice Front')

divider2 = make_axes_locatable(ax)
cax2 = divider2.append_axes("right", size="5%", pad=0.1)
cbar2 = fig.colorbar(img2, cax=cax2)
ticks = np.linspace(-1000, 0, 6)
cbar2.set_ticks(ticks)
cbar2.set_ticklabels([f"{int(abs(t))}" for t in ticks])
cbar2.set_label("Depth (m)", fontsize=12, fontfamily='Times New Roman')
for tick in cbar2.ax.get_yticklabels():
    tick.set_fontname('Times New Roman')
    tick.set_fontsize(12)

ax.set_xlabel('Eastings (m)', fontsize=12, fontfamily='Times New Roman')
ax.set_ylabel('Northings (m)', fontsize=12, fontfamily='Times New Roman')
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(12)
    label.set_family('Times New Roman')
ax.xaxis.get_offset_text().set_fontname('Times New Roman')
ax.xaxis.get_offset_text().set_fontsize(12)
ax.yaxis.get_offset_text().set_fontname('Times New Roman')
ax.yaxis.get_offset_text().set_fontsize(12)
ax.tick_params(direction='in')
ax.set_xlim(xlimit)
ax.set_ylim(ylimit)
ax.set_aspect('equal')

legend = ax.legend(loc='lower right', prop={'size': 11, 'family': 'Times New Roman'})
legend.get_frame().set_linewidth(0)

ax = axes[0]
ax.text(0.03, 0.97, 'a', transform=ax.transAxes, fontsize=12, fontfamily='Times New Roman', verticalalignment='top', horizontalalignment='left', fontweight='bold')
ax = axes[1]
ax.text(0.03, 0.97, 'b', transform=ax.transAxes, fontsize=12, fontfamily='Times New Roman', verticalalignment='top', horizontalalignment='left', fontweight='bold')

plt.subplots_adjust(hspace=0.2)
#plt.savefig("WAOM_BMR_Depth.jpg", dpi=500, bbox_inches='tight')
plt.show()
```
